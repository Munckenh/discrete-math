## Task 1

Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$.

### Proof

To prove that Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$, we can use a proof by contradiction.

Assume for the sake of contradiction that $e \in \mathbb{Q}$. This means that Euler's number can be expressed as a fraction of two positive integers, $\frac{p}{q}$, where $q \ne 0$:

$$\frac{p}{q} = \sum_{k=0}^\infty \frac{1}{k!}$$

By truncating the series after $q$, we get the equation:

$$\frac{p}{q} = \sum_{k=0}^q \frac{1}{k!} + \sum_{k=q+1}^\infty \frac{1}{k!}$$

Multiply both sides of the equation by $q!$. This gives:

$$\begin{align} q! \frac{p}{q} &= q! \sum_{k=0}^q \frac{1}{k!} + q! \sum_{k=q+1}^\infty \frac{1}{k!} \\ p(q-1)! &= \left( \frac{q!}{0!} + \frac{q!}{1!} + \frac{q!}{2!} + \cdots + \frac{q!}{q!} \right) + \left( \frac{q!}{(q+1)!} + \frac{q!}{(q+2)!} + \frac{q!}{(q+3)!} + \cdots \right) \end{align}$$

After tranposing, simplifying, and using the geometric series formula, we have:

$$\begin{align} p(q-1)! - \left( \frac{q!}{0!} + \frac{q!}{1!} + \frac{q!}{2!} + \cdots + \frac{q!}{q!} \right) &= \frac{1}{q+1} + \frac{1}{(q+1)(q+2)} + \frac{1}{(q+1)(q+2)(q+3)} + \cdots \\ &< \sum_{k=1}^\infty \left( \frac{1}{q+1} \right)^k = \frac{\frac{1}{q+1}}{1 - \frac{1}{q+1}} \le 1 \end{align}$$

Notice that the left hand side is an integer. This is a contradiction. We cannot have an integer that is strictly between $0$ and $1$. Therefore, $e$ is irrational.

## Task 2

Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:

$$\left( \sum_{k=1}^n |a_k + b_k|^p \right)^\frac{1}{p} \leq \left( \sum_{k=1}^n |a_k|^p \right)^\frac{1}{p} + \left( \sum_{k=1}^n |b_k|^p \right)^\frac{1}{p}$$

### Proof

Let's prove this using Hölder's inequality. Hölder's inequality states that for any $p>1$ and $\frac{1}{p} + \frac{1}{q} = 1$, where $q$ is the conjugate of the exponent $p$ such that we have:

\begin{align}
\sum_{k=1}^n |a_k + b_k|^p &= \sum_{k=1}^n a_k|a_k+b_k|^{p-1} + \sum_{k=1}^n b_k|a_k+b_k|^{p-1} \\
&\le \left( \sum_{k=1}^n |a_k|^p \right)^\frac{1}{p} \left( \sum_{k=1}^n \left| |a_k+b_k|^{p-1} \right|^q \right)^\frac{1}{q} + \left( \sum_{k=1}^n |b_k|^p \right)^\frac{1}{p} \left( \sum_{k=1}^n \left| |a_k+b_k|^{p-1} \right|^q \right)^\frac{1}{q} \\
&= \left( \left( \sum_{k=1}^n |a_k|^p \right)^\frac{1}{p} + \left( \sum_{k=1}^n |b_k|^p \right)^\frac{1}{p} \right) \left( \sum_{k=1}^n |a_k+b_k|^p \right)^\frac{1}{q}
\end{align}

Divide both sides by:

$$\left( \sum_{k=1}^n |a_k+b_k|^p \right)^\frac{1}{q}$$

Then we get:

$$\left( \sum_{k=1}^n |a_k+b_k|^p \right)^{1-\frac{1}{q}} \le \left( \sum_{k=1}^n |a_k|^p \right)^\frac{1}{p} + \left( \sum_{k=1}^n |b_k|^p \right)^\frac{1}{p}$$

Since $1-\frac{1}{q}=p$:

$$\left( \sum_{k=1}^n |a_k + b_k|^p \right)^\frac{1}{p} \leq \left( \sum_{k=1}^n |a_k|^p \right)^\frac{1}{p} + \left( \sum_{k=1}^n |b_k|^p \right)^\frac{1}{p}$$

## Task 3

Prove the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$.

### Proof

We know that the square of a real number is nonnegative.

$$\begin{align} |x+y|^2 &= (x+y)^2 \\ &= x^2 + 2xy + y^2 \\ &= |x|^2 + 2xy + |y|^2 \end{align}$$

From the negative of absolute value, we know that $-|x| \leq x \leq |x|$; thus,

$$\begin{align} |x+y|^2 &= |x|^2 + 2xy + |y|^2 \\ &\leq |x|^2 + 2|xy| + |y|^2 \\ &\leq |x|^2 + 2|x| \cdot |y| + |y|^2 \\ &= \left( |x| + |y| \right)^2 \end{align}$$

Take the square root of both sides:

$$\begin{align} \sqrt{|x+y|^2} &\leq \sqrt{\left( |x| + |y| \right)^2} \\ |x+y| &\leq |x| + |y| \end{align}$$

## Task 4

Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$.

$$\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}$$

### Proof

Let's use Cauchy-Schwarz inequality which states that for any sequences of real numbers $a_i$ and $b_i$,

$$\left( \sum_{i=1}^n a_i b_i \right)^2 \leq \left( \sum_{i=1}^n a_i^2 \right) \left( \sum_{i=1}^n b_i^2 \right)$$

For our purpose, set $a_i = \frac{u_i}{\sqrt{v_i}}$ and $b_i = \sqrt{v_i}$. Then we have:

$$\begin{align} \left( \sum_{i=1}^n \frac{u_i}{\sqrt{v_i}} \cdot \sqrt{v_i} \right)^2 &\leq \left( \sum_{i=1}^n \left( \frac{u_i}{\sqrt{v_i}} \right)^2 \right) \left( \sum_{i=1}^n \left( \sqrt{v_i} \right)^2 \right) \\ \left( \sum_{i=1}^n u_i \right)^2 &\leq \left( \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \left( \sum_{i=1}^n v_i \right) \end{align}$$

Divide both sides by $\Sigma_{i=1}^n v_i$, we get:

$$\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}$$
